```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_scores.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Drop Unnamed Column

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
df.head()
```

# Gender Distribution

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution Chart")
plt.show()
```

```python
# from the above chart we have analyzed that
# the number of females in the data is more than the number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore" : 'mean',"ReadingScore" : 'mean',"WritingScore" : 'mean',})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb, annot= True)
plt.title("Relation between Parent's Eduaction and Student's Score")
plt.show()
```

```python
# from the above chart we have concluded 
#that the education of the parents have a good impact on the score
```

```python
ab = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean',})
print(ab)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(ab, annot = True)
plt.title("Relation between Parent's Marital Status and Student's Score")
plt.show()
```

```python
# Fom the above chart we have concluded that marital status of parent doesn't impact on the score of students
```

```python
ws = df.groupby("WklyStudyHours").agg({"MathScore" : 'mean',"ReadingScore" : 'mean',"WritingScore" : 'mean',})
print(ws)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(ws, annot = True)
plt.title("Relation between Weekly Study Hours and Student's Score")
plt.show()
```

```python
# From the above chart we have concluded that there is little bit differnce in marks according to weekly study hours
```
